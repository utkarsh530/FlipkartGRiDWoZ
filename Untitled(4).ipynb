```python
speech = os.listdir("speech")
```

```python
noise = df["slice_file_name"].values
```

```python
from pydub import AudioSegment
import random
```

```python
j = 0
for s in speech:
    sound1 = AudioSegment.from_file("speech/"+s)
    for i in range(10):
        idx = random.randint(0, len(noise)-1)
        n = noise[idx]
        sound2 = AudioSegment.from_file("Urban/"+n)
        start = random.randint(0, 2500)
        combined = sound1.overlay(sound2, loop=True, position = start)
        combined.export("combined_street/"+str(j)+".wav", format='wav')
        sound1.export("parallel_speech/"+str(j)+".wav",  format='wav')
        print(j)
        j = j + 1
```

```python
combined = os.listdir("combined_street")
```

```python
n_fft = 255
hop_length_fft = 63
frame_length = 8064
k = 0
for s in combined:
    i = 0
    j = 0
    try:
        audio,sr = librosa.load("combined_street/"+s)
        speech,sr = librosa.load("parallel_speech/"+s)
        while (i+frame_length<len(audio)):
            stftaudio = librosa.stft(audio[i:i+frame_length], n_fft=n_fft, hop_length=hop_length_fft)
            stftaudio_magnitude, stftaudio_phase = librosa.magphase(stftaudio)

            stftaudio_magnitude_db = librosa.amplitude_to_db(
                stftaudio_magnitude, ref=np.max)

            np.save("numpy_combined/"+str(k), stftaudio_magnitude_db)

            stftaudio = librosa.stft(speech[j:j+frame_length], n_fft=n_fft, hop_length=hop_length_fft)
            stftaudio_magnitude, stftaudio_phase = librosa.magphase(stftaudio)

            stftaudio_magnitude_db = librosa.amplitude_to_db(
                stftaudio_magnitude, ref=np.max)

            np.save("numpy_speech/"+str(k), stftaudio_magnitude_db)

            i += frame_length
            j += frame_length
            k += 1
            print(k)
    except:
        pass
```
