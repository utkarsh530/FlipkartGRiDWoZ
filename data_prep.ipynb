```
from pydub import AudioSegment
```

```
sound1 = AudioSegment.from_file("speech.wav")
sound2 = AudioSegment.from_file("noise.wav")

combined = sound1.overlay(sound2, times=200)

combined.export("combined.wav", format='wav')
```
